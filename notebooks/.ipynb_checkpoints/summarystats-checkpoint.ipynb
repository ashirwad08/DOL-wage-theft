Location Severity Quotient = Severity of Violation in MSA / Severity of Violation Statewide

```python
#1) For MSA, for NAICS3: MSA Severity Min Wage Cases = MinWage_ATP * MinWage_EmpAff / EMPL_N
#2) Statewide, for NAICS3: Severity Min Wage Cases = Sum(MinWage_ATP)*Sum(MinWage_EmpAff)/Sum(EMPL_N)

#MSA Severity Quotient for NAICS3 = 1/2


grp_naics3=dat_ca_msa.groupby(by='NAICS3_x')['All_AtpAmt','ee_violtd_cnt','EMPL_N'].sum()
```

```python
grp_naics3.reset_index(inplace=True)
```

```python
#merge it back to main so that all NAICS3s have the same aggregate #s
dat_ca_msa1=dat_ca_msa.merge(grp_naics3, how='inner', left_on='NAICS3_x', right_on='NAICS3_x')
```

```python
dat_ca_msa1['Severity_AllVltns_MSA'] = dat_ca_msa1['ee_violtd_cnt_x']*dat_ca_msa1['All_AtpAmt_x']/dat_ca_msa1['EMPL_N_x']
dat_ca_msa1['Severity_AllVltns_CA'] = dat_ca_msa1['ee_violtd_cnt_y']*dat_ca_msa1['All_AtpAmt_y']/dat_ca_msa1['EMPL_N_y']
dat_ca_msa1['MSA_SeverityQuotient'] = dat_ca_msa1['Severity_AllVltns_MSA']/dat_ca_msa1['Severity_AllVltns_CA']
```

```python
pd.set_option('max_columns',50)
dat_ca_msa1.sample(2)
```

```python
#similarity
mat = np.matrix(dat_nonull[cluster_cols])
```

```python
mat.shape
```

```python
from sklearn.preprocessing import normalize
mat_norm = normalize(mat, axis=0)
#similarity
from scipy.spatial import distance
sq_mat = distance.squareform(distance.pdist(mat_norm, 'minkowski',p=14))
```

```python
#function to take in one node (MSA industry) and return all similar nodes 

def single_node_json(mat, labels, sourceIndex, groups):
    links=[]
    nodes=[]
    for i in range(mat.shape[0]):
        #source = i[0]
        target = i
        val = mat[sourceIndex, target]

        if val<=0.001:
            links.append({'source':sourceIndex, 'target':target, 'value':val*1000})
            #nodes.append({'name':labels[i], 'group':groups[i]})
        
        
    #nodes=[]
    for i in range(mat.shape[0]):
        #nodes.append({'name':ntnl_flsa_summ['naics_code_description'][i], 'group':np.random.randint(low=1,high=5)})
        nodes.append({'name':labels[i], 'group':groups[i]})
    
    graphdict={'nodes':nodes, 'links':links}
    return graphdict
```
