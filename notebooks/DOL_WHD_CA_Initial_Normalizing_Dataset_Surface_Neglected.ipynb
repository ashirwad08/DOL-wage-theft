```python
#plot NAICS Desc. Frequency
fig = plt.figure(figsize=(15,5))
whd_enf['NAICS Description'].value_counts().head(50).plot(kind='bar')
```

```python
whd_enf[whd_enf['NAICS Description']=="Women's, Girls', and Infants' Cut and Sew Apparel Contractors"]['Total Case Violations'].sum()
```

```python
#Remember, the dataset is self serving. We have more data BECAUSE we know these industries violate more. 
# Gotta think about the ones with low reported cases but high violations. Let's take a ratio of #violations and frequency

# So let's create a new metric: Total Violations/Enforced Cases (Enforcement Payoff?)
Enf_Payoff=(whd_enf.groupby(by='NAICS Description')['Total Case Violations'].sum()/whd_enf.groupby(by='NAICS Description').size()).reset_index()
```

```python
Enf_Payoff.columns = ['NAICS Description','Violations Per Enforced Case']
```

```python
Enf_Payoff.index=Enf_Payoff['NAICS Description']
```

```python
# Enforcement really paid off in these industries!!
# Small number of observations, most likely, but the dataset is self serving by Michael's admission 
#(more cases enforced because we know these industries are violators)
# So, which industries are "similar" to these below?? Maybe get other NAICS codes not in dataset and get a dot prod?

ax1 = fig.add_subplot(111)
Enf_Payoff.sort_values(by='Violations Per Enforced Case', ascending=False).head(50).plot(kind='bar',
               y='Violations Per Enforced Case',figsize=(20,5))
plt.ylim([0,3500])
ax1.set_ylabel('Enforcement Payoff\n(Total Case Violations/Enforced Case)')

ax2 = ax1.twinx()
Enf_Payoff.sort_values(by='Violations Per Enforced Case', ascending=False).head(50).plot(kind='bar',
                                                                                        y='Enforced_Cases')
ax2.set_ylabel('Actual Cases Enforced')
```

```python
temp1=pd.DataFrame(index=whd_enf['NAICS Description'].value_counts().index)
temp1['Enforced_Cases']=whd_enf['NAICS Description'].value_counts()
temp1['Enforced_Prcnt']=whd_enf['NAICS Description'].value_counts(normalize=True)
temp1.head()
```

```python
Enf_Payoff=Enf_Payoff.merge(temp1, how='inner', left_index=True, right_index=True)
```

```python
Enf_Payoff.sort_values(by='Violations Per Enforced Case', ascending=False)['Violations Per Enforced Case'].head(50)
```

```python
#ypos = np.arange(len(Enf_Payoff.sort_values(by='Violations Per Enforced Case', ascending=False).head(50).index))

fig = plt.figure(figsize=(20,15))
ax1 = fig.add_subplot(211)
Enf_Payoff.sort_values(by='Violations Per Enforced Case', ascending=False)['Violations Per Enforced Case'].head(50).plot(kind='bar',
                                                                                                                        facecolor='green')   
#plt.xticks(ypos, Enf_Payoff.sort_values(by='Violations Per Enforced Case', ascending=False).head(50).index)
plt.ylim([0,3500])
ax1.set_ylabel('Enforcement Payoff\n(Total Case Violations/Enforced Case)')
ax1.set_title('Understading "Enforcement Payoff" in terms of\nTotal Violations Discovered')
ax2 = fig.add_subplot(212)
#ax2=ax1.twinx()
Enf_Payoff.sort_values(by='Violations Per Enforced Case', ascending=False)['Enforced_Cases'].head(50).plot(kind='bar', 
                                                                                                           facecolor='orange', 
                                                                                                           alpha=0.95)
#plt.ylim([0,400])
ax2.set_ylabel('Number of Cases Enforced')

plt.subplots_adjust(wspace=0, hspace=0)
```

```python
#this makes it seem like industries with just 1 observation have high payoff, whereas those with a few more observations 
#have less payoff comparitively. 
#Let's correct for that by throwing in a logarithmic smoother that rewards a higher number of cases (that consistently have somewhat high violations)
# i.e. Enforcement Payoff = (Total Violations/Enforced Cases) * ln(Enforced Cases+1)

Enf_Payoff['ViolationRate_Smoothed'] = Enf_Payoff['Violations Per Enforced Case']*np.log(Enf_Payoff['Enforced_Cases']+1)
```

```python
Enf_Payoff.sort_values(by='Enforced_Cases', ascending=False)['Enforced_Cases'].head(50)
```

```python
#now, take a fresh look at the plot to understand our "Enforcement Payoff" score

fig = plt.figure(figsize=(20,15))
ax1 = fig.add_subplot(211)
Enf_Payoff.sort_values(by='ViolationRate_Smoothed', ascending=False)['ViolationRate_Smoothed'].head(50).plot(kind='bar',
                                                                                                                        facecolor='green')   
#plt.xticks(ypos, Enf_Payoff.sort_values(by='Violations Per Enforced Case', ascending=False).head(50).index)
#plt.ylim([0,3500])
ax1.set_ylabel('Enforcement Payoff Score\n(Adjusted for num. observations)')
ax1.set_title('Understading "Enforcement Payoff" in terms of\nTotal Violations Discovered\n(Enforcment Payoff Decreases from Left to Right)')
ax2 = fig.add_subplot(212)
#ax2=ax1.twinx()
Enf_Payoff.sort_values(by='ViolationRate_Smoothed', ascending=False)['Enforced_Cases'].head(50).plot(kind='bar', 
                                                                                                           facecolor='orange', 
                                                                                                           alpha=0.95)
#plt.ylim([0,400])
ax2.set_ylabel('Number of Cases Enforced')

plt.subplots_adjust(wspace=0, hspace=0)
```

```python
#finally, if we treat teh industries with less than 5 reported cases as not having enough evidence (outliers), 
#then this is how our chart looks
#now, take a fresh look at the plot to understand our "Enforcement Payoff" score

fig = plt.figure(figsize=(20,15))
ax1 = fig.add_subplot(211)
Enf_Payoff.loc[Enf_Payoff['Enforced_Cases']>=5,:].sort_values(by='ViolationRate_Smoothed', ascending=False)['ViolationRate_Smoothed'].head(50).plot(kind='bar',
                                                                                                                        facecolor='green')   
#plt.xticks(ypos, Enf_Payoff.sort_values(by='Violations Per Enforced Case', ascending=False).head(50).index)
#plt.ylim([0,3500])
ax1.set_ylabel('Enforcement Payoff Score\n(Adjusted for num. observations)')
ax1.set_title('Understading "Enforcement Payoff" in terms of\nTotal Violations Discovered\n(Enforcment Payoff Decreases from Left to Right)')
ax2 = fig.add_subplot(212)
#ax2=ax1.twinx()
Enf_Payoff.loc[Enf_Payoff['Enforced_Cases']>=5,:].sort_values(by='ViolationRate_Smoothed', ascending=False)['Enforced_Cases'].head(50).plot(kind='bar', 
                                                                                                           facecolor='orange', 
                                                                                                           alpha=0.95)
#plt.ylim([0,400])
ax2.set_ylabel('Number of Cases Enforced\n(at least 5 to qualify)')

plt.subplots_adjust(wspace=0, hspace=0)
```
